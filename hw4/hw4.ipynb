```python
import numpy as np

def gen_rho_sigma(alpha):
    a_0 = alpha 
    a_1 = -1-alpha
    A = [[2,1],[2,1/2]]
    b = [2+(a_0/2),(4/3)+(a_0/6)]
    b_n1,b_0 = np.linalg.solve(A,b)
    b_1 = 2+a_0-b_0-b_n1
    # print(b_n1,b_0)
    # roots = np.roots([1,a_0,a_1])
    
    def rho(z):

        return z**2 + a_0*z + a_1

    def sigma(z):
        return b_n1*z**2 + b_0*z + b_1
    print('\\alpha = ',alpha)
    # print('\\rho(z_1) = ',rho(roots[0]))
    # print('|z_1| = ',np.abs(roots[0]))
    # print('\\rho(z_2) = ',rho(roots[1]))
    # print('|z_2| = ',np.abs(roots[1]))
    return rho,sigma
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,  # Use LaTeX for text rendering
    "font.family": "serif",  # Set the font family to serif
    "font.serif": ["Times New Roman"],  # Specify the serif font
    "font.size": 22,  # Set the default font size
})

fig,ax = plt.subplots(figsize=(8,8))

# for alpha in [-1.8+.1*i for i in range(8)]:
# for alpha in [-1.0+.1*i for i in range(10)]:
for alpha in np.round(np.linspace(-1.8, -1.1, num=8),decimals=1):
# for alpha in np.round(np.linspace(-1.0, -0.1, num=10),decimals=1):
    # print(alpha)
    rho,sigma = gen_rho_sigma(alpha=alpha)
    theta = np.linspace(0,2*np.pi,1000,dtype=np.float128)
    # print(rho(np.exp(1j*theta)),)
    out = rho(np.exp(1j*theta))/sigma(np.exp(1j*theta))
    # ax.scatter([np.mean(np.real(out)),],[np.mean(np.imag(out))],marker='x')
    pt = np.mean(out)
    print(np.abs(rho(pt)))
    ax.scatter([np.real(pt),],[np.imag(pt)],marker='x')
    ax.plot(np.real(out),np.imag(out),label=f'$\\alpha = {alpha}$')
    ax.grid(visible=True)

ax.set_xlabel(r'$\textsf{Re}(z)$')
ax.set_ylabel(r'$\textsf{Im}(z)$')
plt.legend(fontsize=16)
plt.show()
```
